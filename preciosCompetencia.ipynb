```python
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Precios_Competencia.xlsx')
```

```python
df.set_index('Unnamed: 0', inplace=True)
```

```python
# rename index
df = df.rename_axis('Competencia')
df = df.sort_values(by='$/mL', ascending=False)
```

```python
default = 'silver'

colores = [default, default, default, default, default, 
          default, default, default, default, 
          'lime', default, default, default]

plt.figure(figsize=(22,10))
plt.title('Precio Por mL en Bebidas RTD')
plt.xlabel('RTD')
plt.ylabel('$ / mL')
plt.bar(list(df.index), df['$/mL'], color=colores)
for i in range(len(df)):
    plt.text(i, df['$/mL'][i], df['$/mL'][i], ha = 'center', va = 'bottom', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
df_aloe = df.drop(['Frutto Litteral', 'Hatsu Te', 'Hatsu Soda', 'Mr.Tea Light', 'Brisa', 'Red Bull Sin Azúcar'], axis=0)
```

```python
default = 'silver'

colores = [default, default, default, default, 
          default, 'lime', default]

plt.figure(figsize=(20,10))
plt.title('Precio Por mL en Bebidas Aloe RTD')
plt.xlabel('RTD')
plt.ylabel('$ / mL')
plt.bar(list(df_aloe.index), df_aloe['$/mL'], color=colores)
for i in range(len(df_aloe)):
    plt.text(i, df_aloe['$/mL'][i], df_aloe['$/mL'][i], ha = 'center', va = 'bottom', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```
